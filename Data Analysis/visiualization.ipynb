```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,  2*np.pi , 50)
x = np.sin(t)
y = np.cos(t)
```

```python
plt.figure(figsize=(15,7))
plt.plot(t,x,'bo--')
plt.plot(t,y,'H--', markersize=10) 
```
